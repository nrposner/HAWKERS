```python
#revised version

def corpus_extractor():
    """Pulls out the ninth column from the dataset in order to get
    the raw corpus which will be use din preprocessing
    """
    data = pd.read_csv("service_reviews_15000rows_translated.csv")
    corpus = data.iloc[:, 8]
    return corpus

def preprocess(corpus:pd.core.series.Series, min_len:int = 3, max_len:int = 15) -> list:
    """ Take in a corpus of text in a pandas series and perform
    preprocessing

    corpus: a pandas series containing text

    min_len: minimum word length. No shorter words will be retained

    max_len: maximum word length. No longer words will be retained
    """
    
    if not (min_len <= max_len):
        raise ValueError("make sure your minimum and maximum token lengths are not reversed")

    preprocessed_corpus = []

    for i in corpus:
        preprocessed_doc = gensim.utils.simple_preprocess(i, min_len = min_len, max_len = max_len)
    
        preprocessed_corpus.append(preprocessed_doc)

        # go line by line, removing common words
    stoplist = set('for a of the and to in'.split(' '))
    texts = [[word for word in document if word not in stoplist]
         for document in preprocessed_corpus]

    # count word frequencies
    frequency = defaultdict(int)
    for text in texts:
        for token in text:
            frequency[token] += 1

    # only keep words that appear more than once
    processed_corpus = [[token for token in text if frequency[token] > 1] for text in texts]

    return processed_corpus

def corpus_maker(processed_corpus:list):
    """ Take in a processed corpus from preprocessing and transform it into
    a tfidf bag of words corpus
    
    """
    # turn this into a dictionary structure
    dictionary = corpora.Dictionary(processed_corpus)

    # create a 'bag of words' corpus using that dictionary
    bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]

    # train the model

    # tfidf is a transformation that finds term frequency in model frequency
    # we will use this in order to create a structure which other models can attack more easily
    tfidf = models.TfidfModel(bow_corpus)

    corpus_tfidf = tfidf[bow_corpus]

    return corpus_tfidf, dictionary

#can we perhaps improve performance by training the model outside similarity_order?? 
#no reason to iterate above the model definition, we just lose performance on it

def similarity_calculator_k(corpus, corpus_tfidf: gensim.interfaces.TransformedCorpus, dictionary: gensim.corpora.dictionary.Dictionary, query_dict:dict, mod, num_topics:int, k:int):
    """ Take in a tfidf corpus and dictionary created by corpus_creation,
    as well as multiple classified queries and a model. For each query we then get a
    similarity score for each document, and count how many of the top k similarity 
    scores belonged to each category: finally, we select the highest-performing category
    and classify the document accordingly. 
    This remains a naive classifier, more refinement is needed. LSI model is recommended

    mod: must be formatted as models.ModelName, such as models.LdaModel, or models.LsiModel
    """
    
    #or should we do it in parallel? none of this monkeying around with dfs
    
    df_list = []
    
    for category in query_dict:
        queries = query_dict[category]
        
        #every column in the dataframe is a query
        df = pd.DataFrame(columns = queries)
        
        for q in queries:
            
            vec_bow = dictionary.doc2bow(q.lower().split())
            vec_model = mod[vec_bow]
            
            #index these
            index = similarities.MatrixSimilarity(mod[corpus_tfidf])
            
            sims = index[vec_model]
            
            df[q] = sims
        
        df_list.append(df)
        
    
    #now going through these dataframes
    
    
    classifications = []
    
    
    for i in np.arange(len(corpus_tfidf)):
        
        scores = []
        classes = []
        
#         sort_df = pd.DataFrame(columns = ["scores", "classes"])
        
        #better to make into one with tags and order?
        for df, class_name in zip(df_list, list(query_dict.keys())): #can we streamline this with pandas methods? prob not, it's only a few dfs anyway
            
            scores.append(df.iloc[i].values)
            classes.append(class_name)
            
#             temp_df = pd.DataFrame(columns = ["scores", "classes"])
            
#             temp_df["scores"] = df.iloc[i]
#             temp_df["classes"] = class_name
            
            
            
#             pd.concat(sort_df, temp_df)
            
#             sort_df.append(temp_df)
        
        
        sort_df = pd.DataFrame({"scores":scores, "classes":classes})
        
#         sort_df.dropna(how = "any", inplace = True)
        
#         #error, need to remove nans?
#         sort_df.sort_index(inplace = True)
        
        print(i)
        
        sort_df.sort_values(by="scores", inplace=True)
        
        classification = sort_df["classes"][0:k].mode()
        
        classifications.append(classification)
    
    classifications_df = pd.DataFrame({"documents":train_corpus, "class": classifications})
    
    return classifications_df
    


def classification_pipeline_k(train_corpus:pd.Series, query_dict:dict, mod, num_topics:int, k:int):
    
    #query sets should bea dictionary of lists
    
    processed_corpus = preprocess(train_corpus)
    corpus_tfidf, dictionary = corpus_maker(processed_corpus)
    df = pd.DataFrame(columns = list(query_dict.keys())) ##check
    
    model = mod(corpus_tfidf, id2word=dictionary, num_topics=num_topics)
    
    return similarity_calculator_k(train_corpus, corpus_tfidf, dictionary, query_dict, model, 4, 10)
    
    
    
```
